## Validación de la métrica de ventas

Se realizó una validación comparando el valor de ventas reportado (`SALES`) con el cálculo basado en la cantidad y el precio unitario.

Durante este análisis se identificaron diferencias en varios registros. Algunas de estas diferencias son mínimas y se deben a temas de precisión decimal, sin embargo, también se encontraron diferencias más significativas.

Esto sugiere que el valor de ventas puede estar influenciado por factores adicionales como descuentos, impuestos u otros ajustes que no están explícitos en el dataset.

Por esta razón, se decidió mantener la columna `SALES` como fuente principal para el análisis, ya que representa el valor final de la transacción.

## Asignación de clientes a ventas

Después de normalizar los nombres de los países, se identificó que no existe una correspondencia completa entre los países de los datasets de ventas y clientes.

Debido a esta limitación, se optó por asignar clientes de forma aleatoria a cada venta.

Esta decisión permite mantener la estructura del modelo sin asumir relaciones incorrectas entre clientes y ubicaciones geográficas.

```python
cols = [
    'ORDERNUMBER',
    'ORDERDATE',
    'product_id',
    'customer_id',
    'QUANTITYORDERED',
    'PRICEEACH',
    'SALES',
    'COUNTRY'
]

sales_final = sales[cols].copy()
```

```python
import requests

response = requests.get("https://restcountries.com/v3.1/all")
countries_data = response.json()

regions_list = []
for c in countries_data:
    regions_list.append({
        "country_name": c.get("name", {}).get("common", ""),
        "region": c.get("region", ""),
        "subregion": c.get("subregion", ""),
        "continent": c.get("continents", [""])[0]
    })

df_regions = pd.DataFrame(regions_list)
df_regions.head()
```

```python
response = requests.get("https://restcountries.com/v3.1/all?fields=name,region,subregion,continents")
print(response.status_code)
countries_data = response.json()
print(countries_data[0])
```

```python
regions_list = []
for c in countries_data:
    regions_list.append({
        "country": c["name"]["common"],
        "region": c.get("region", ""),
        "subregion": c.get("subregion", ""),
        "continent": c["continents"][0]
    })

df_regions = pd.DataFrame(regions_list)

# Relacionar con ventas
sales_final = sales_final.merge(
    df_regions[["country", "region", "subregion", "continent"]],
    on="country",
    how="left"
)

# Verificar nulls (países que no hicieron match)
print("Sin match:", sales_final[sales_final["region"].isna()]["country"].unique())
```

```python
# Crear dimensión regions
regions_final = sales_final[["country", "region", "subregion", "continent"]].drop_duplicates().reset_index(drop=True)
regions_final.insert(0, "region_id", regions_final.index + 1)

# Agregar region_id a sales_final
sales_final = sales_final.merge(regions_final[["region_id", "country"]], on="country", how="left")

# Guardar los 4 CSVs
regions_final.to_csv("../data/processed/regions.csv", index=False)
sales_final.to_csv("../data/processed/sales.csv", index=False)

print(f"✅ {len(regions_final)} regiones")
print(f"✅ sales_final columnas: {list(sales_final.columns)}")
print(regions_final)
```

```python
# Dimensión tiempo extraída de order_date
time_final = sales_final[["order_date"]].drop_duplicates().reset_index(drop=True)
time_final.insert(0, "time_id", time_final.index + 1)

time_final["year"]      = pd.to_datetime(time_final["order_date"]).dt.year
time_final["month"]     = pd.to_datetime(time_final["order_date"]).dt.month
time_final["month_name"]= pd.to_datetime(time_final["order_date"]).dt.strftime("%B")
time_final["quarter"]   = pd.to_datetime(time_final["order_date"]).dt.quarter
time_final["day"]       = pd.to_datetime(time_final["order_date"]).dt.day
time_final["weekday"]   = pd.to_datetime(time_final["order_date"]).dt.strftime("%A")

time_final.to_csv("../data/processed/time.csv", index=False)
print(f"✅ {len(time_final)} fechas únicas")
print(time_final.head())
```

```python
sales_final = sales_final.merge(time_final[["time_id", "order_date"]], on="order_date", how="left")
sales_final.to_csv("../data/processed/sales.csv", index=False)
print(list(sales_final.columns))
```

```python
from dotenv import load_dotenv
from sqlalchemy import create_engine
import os

load_dotenv()

engine = create_engine(
    f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)

print("✅ Conexión creada")
```

```python
products_final.to_sql("products", engine, if_exists="append", index=False)
customers_final.to_sql("customers", engine, if_exists="append", index=False)
regions_final.to_sql("regions", engine, if_exists="append", index=False)
time_final.to_sql("time", engine, if_exists="append", index=False)
sales_final.to_sql("sales", engine, if_exists="append", index=False)

print("✅ 5 tablas cargadas en PostgreSQL")
```

```python
# Limpiar tablas en orden inverso (primero la de hechos)
with engine.connect() as conn:
    conn.execute("TRUNCATE TABLE sales, time, regions, customers, products RESTART IDENTITY CASCADE")
    conn.commit()

print("✅ Tablas limpiadas")
```
